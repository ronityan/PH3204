```python
import pandas as pd

df=pd.read_csv(r"D:\ronit\IISER-K\6 SEM\PH3204\ph3204\expt01\data\part01_01.csv")

df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = df.columns[3]
y = df.columns[1]

filtered_df = df[(df[x] > 0.18) & (df[y] > 0)]  

x_data = filtered_df[x].values
y_data = filtered_df[y].values

def linear_model(x, a, b):
    return a * x + b

params, covariance = curve_fit(linear_model, x_data, y_data)
slope, intercept = params  

x_fit = np.linspace(min(x_data), max(x_data), 100) 
y_fit = linear_model(x_fit, *params)

plt.scatter(x_data, y_data, color='green', label="Data")

plt.plot(x_fit, y_fit, color='blue', linestyle='--', label=rf"$I_s = {slope:.2f}I_z + {intercept:.2f}$")

plt.xlabel(r"$I_z$ (mA)")
plt.ylabel(r"$I_s $(mA)")
plt.grid()
plt.legend()

#Show the plot
plt.savefig("part01_01.png")
plt.show()
covariance
```
